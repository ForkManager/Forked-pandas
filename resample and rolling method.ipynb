```python
df['Adj Close'].plot()
# df['Adj Close'].resample('W').mean().plot()
# df['Adj Close'].resample('M').mean().plot()
df['Adj Close'].resample('Q').mean().plot()
# df['Adj Close'].resample('Y').mean().plot()
```

```python
# df['Adj Close'].resample('W').std().plot()
df['Adj Close'].resample('M').std().plot()
# df['Adj Close'].resample('Q').std().plot()
# df['Adj Close'].resample('Y').std().plot()
```

```python
df['Adj Close']['2017-01-01':'2017-03-31'].plot()
df['Adj Close']['2017-01-01':'2017-03-31'].rolling(21).mean().plot()
df['Adj Close']['2017-01-01':'2017-03-31'].resample('M').mean().plot()
```

```python
df['Adj Close']['2017-01-01':'2017-03-31'].rolling(21).std().plot()
df['Adj Close']['2017-01-01':'2017-03-31'].resample('M').std().plot()
```
